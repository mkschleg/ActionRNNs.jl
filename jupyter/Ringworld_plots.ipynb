```julia
using Revise;
using Reproduce
using ProgressMeter
includet("../plot/ringworld.jl")
```

```julia

clean_data=(data)->data["results"]["end"]
ic = Reproduce.ItemCollection("../data/ringworld_lstm_rmsprop/data/")

o_plts = []

for outhorde ∈ diff(ic)["outhorde"]
    plts = []
    for rnn_config ∈ diff(ic)["rnn_config"]
        plt = heatmap_ring(
            ic, 
            "numhidden", 
            "truncation", 
            "alpha"; 
            static_params=Dict{String, Any}("outhorde"=>outhorde, "rnn_config"=>rnn_config), 
            clean_data=clean_data,
            title=outhorde*"-"*rnn_config, clim=(0.05, 0.4))
        push!(plts, plt)
    end
    push!(o_plts, plot(plts...; layout=(1,3)))
end

plot(o_plts...; layout=(2,1), size=(900, 400))
```
